```python
def summ_l(l):
    s=0
    for i in l:
        s+=i
    return s
l = [10,11,14,8]
print(summ_l(l))
```

```python


def max_l(l):
    lar = l[0]
    for i in l:
        if i> lar:
            lar=i
    return lar
l = [10,11,14,8]
print(max_l(l)*2)
```
